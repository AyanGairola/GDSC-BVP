```python
import pandas as pd 
import numpy as np  
import matplotlib.pyplot as plt                                       
from sklearn.model_selection import train_test_split, cross_val_score 
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay, classification_report 
```

```python
df = pd.read_csv('./data.csv')
df
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
df.isnull().values.any()
```

# one hot encoding categorical values

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
categorical_cols = df.select_dtypes(include=['object']).columns
categorical_cols
```

```python
encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')
```

```python
encoded_cols = encoder.fit_transform(df[categorical_cols])
encoded_cols
```

```python
encoded_df = pd.DataFrame(encoded_cols, columns=encoder.get_feature_names(categorical_cols))
encoded_df
```

```python
df = df.drop(categorical_cols, axis=1)
df
```

```python
df = pd.concat([df, encoded_df], axis=1)
```

```python

df
```

```python
df = df.dropna()
```

```python
df
```

```python
dataset = df.to_numpy()
print(dataset.shape)
```

```python
X = dataset[:,:31]
print(X)
print(X.shape)
```

```python
y = dataset[:,31:]
print(y)
print(y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
print("train data :", X_train.shape)
print("\ntest data :", X_test.shape)
print("\ntrain target :", y_train.shape)
print("\ntest target :",y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,  Dropout
from tensorflow.keras import regularizers
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model = Sequential()
model.add(Dense(64, activation='relu', kernel_regularizer=regularizers.l2(0.01), input_shape=(X_train_scaled.shape[1],)))
model.add(Dropout(0.4))
model.add(Dense(64, activation='relu', kernel_regularizer=regularizers.l2(0.01)))
model.add(Dropout(0.4))
model.add(Dense(4, activation='softmax'))
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train_scaled, y_train, epochs=10, batch_size=32, validation_data=(X_test_scaled, y_test))
```

```python
y_pred = model.predict(X_test_scaled)
y_pred_classes = np.argmax(y_pred, axis=1)
y_test_classes = np.argmax(np.array(y_test), axis=1)

# Calculate the accuracy score
accuracy = accuracy_score(y_test_classes, y_pred_classes)
print(f'Accuracy: {accuracy}')
```
